```python
def divide_work(worklist, n):
    chunks = []
    for i in range(0, len(worklist), n):
        chunks.append(worklist[i:i+n])
    return chunks

def copy_directory(src_dir, dest_dir):
    try:
        shutil.copytree(src_dir, dest_dir)
        print(f"Successfully copied {src_dir} to {dest_dir}")
    except Exception as e:
        print(f"Error copying {src_dir}: {str(e)}")

def copy_subdirectory(subdir):
    src_dir = os.path.join(base_dir, subdir)
    dest_dir = os.path.join(output_base_dir, subdir)
    copy_directory(src_dir, dest_dir)

if __name__ == "__main__":
    base_dir = train_cut_audio_  # Change this to the source directory path
    output_base_dir = strain_cut_audio_  # Change this to the destination directory path

    # Ensure the destination directory exists or create it
    if not os.path.exists(output_base_dir):
        os.makedirs(output_base_dir)

    subdirectories = [d for d in os.listdir(base_dir) if os.path.isdir(os.path.join(base_dir, d))]

    divided_worklist = divide_work(subdirectories, cpu_count())

    for idx, workchunk in enumerate(divided_worklist): 
        print(f"Group {idx}")
        pool = Pool(processes=32)

        for i, subdir in enumerate(workchunk):
            print(f"Start {subdir}")
            result = pool.apply_async(copy_directory, args=(os.path.join(base_dir, subdir), os.path.join(output_base_dir, subdir)))
        pool.close()
        pool.join()

    # # Adjust the number of processes as needed
    # num_processes = len(subdirectories)  # You can change the number of processes as needed

    # with Pool(processes=num_processes) as pool:
    #     pool.map(copy_subdirectory, subdirectories)
```
